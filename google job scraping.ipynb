```python
import selenium
import requests
import numpy as np
import pandas as pd
from PIL import Image
from io import BytesIO
from bs4 import BeautifulSoup
from selenium import webdriver
```

```python

class jobs_intern(object):
    def internship(self):
        '''
        This function scrape google job site and extract informations like company name , job title , location ,
        posted date , salary and company logos
    
        Parameters
        ----------
        url is given inside the function itself 

        Returns
        -------
        df - Dataframe containing all the above mention details
        csv file - csv file containg all the data present in the dataframe
   
        '''
        #we have install the firefox driver for the working of selenium
        # https://github.com/mozilla/geckodriver/releases -- use the link install firefox driver
        path_driver = r"C:\Users\user p c\Desktop\python\geckodriver"
        browser = webdriver.Firefox(executable_path = path_driver)
        
        #first url is given and getting the content and extract the original url
        url = f"https://www.google.com/search?q=jobs+near+me"
        browser.get(url)
        soup = BeautifulSoup(browser.page_source,features = "lxml")
        web_url = soup.find("a",{"class":"esVihe"})
        main_url = str(web_url.get("href"))
        
        
        #getting the content from the original website
        browser.get(main_url)
        soup1 = BeautifulSoup(browser.page_source, "lxml" )
        jobs = soup1.find_all("div",{"class":"PwjeAc"})



        # creating the header for csv file
        header = ["COMPANY NAME","JOB TITLE","LOCATION","POSTED DATE","SALARY","LOGOS"]
        
        # creating empty list for appending data
        company_name = []
        job_title = []
        location = []
        posted_date = []
        salary = []
        img = []
        
        # extracting the data
        for job in jobs:
            company_name.append(job.find("div",{"class":"vNEEBe"}).text)
            job_title.append(job.find("div",{"class":"BjJfJf PUpOsf"}).text)

            location.append(job.find("div",{"class":"Qk80Jf"}).text)
            post = job.find("span",{"class":"LL4CDc"}).text
            if post.endswith("ago")== True:
                posted_date.append(post)
            else:
                posted_date.append("Not Given")
            sal = job.find("span",{"class":"LL4CDc"}).text
            if sal.startswith("₹")== True:
                salary.append(sal)
            else:
                salary.append("Not Given")

            logo_link = soup1.find("img",{"class":"YQ4gaf zr758c"})
            images = logo_link.get("src")
            response = requests.get(images)
            img.append(np.array(Image.open(BytesIO(response.content))))

        
        # creating a dictionary with the data 
        df_dict = {"COMPANY NAME":company_name,"JOB TITLE":job_title,"LOCATION":location,"POSTED DATE":posted_date,"SALARY":salary,"logos":img}
        
        # creating a dataframe
        df = pd.DataFrame(df_dict)
        
        # creating the  csv file from dataframe
        df.to_csv('google_job.csv')
        return df
```
